```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set_style("darkgrid")
income = [100, 80, 150, 48, 52, 69, 88]
expense = [30, 100, 100, 20, 75, 50, 28]
index = ['A', 'B', 'C','D', 'E', 'F', 'G']
df = pd.DataFrame({'income': income,'expense': expense}, index=index)
df.head()
```

```python
fig, ax = plt.subplots()

p = np.arange(len(index))
width = 0.2

ax.bar(p - 0.1, income, width, label="income")
ax.bar(p + 0.1, expense,   width, label="expense")

ax.set_xticks(p)
ax.set_xticklabels(index,rotation=90)

plt.legend()
plt.show()
```

```python
fig, ax = plt.subplots()

p = np.arange(len(index))
width = 0.3

ax.bar(p, income, width, label="income")
ax.bar(p, expense, width, bottom = df["income"], label="expense")

ax.set_xticks(p)
ax.set_xticklabels(index,rotation=90)


plt.legend()
plt.show()
```

```python
#3. Rotate the labels and set figsize
fig, ax = plt.subplots(figsize=(8,5))

p = np.arange(len(index))
width = 0.5

ax.bar(p, income, width, label="income")
ax.bar(p, expense, width, bottom = df["income"], label="expense")

ax.set_xticks(p)
ax.set_xticklabels(index)

plt.legend()
plt.show()
```
